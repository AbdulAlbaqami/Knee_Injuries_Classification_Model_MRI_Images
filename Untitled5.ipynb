```python
import os, importlib, sys
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
from torchvision.models import resnet18, ResNet18_Weights
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
from sklearn.metrics import roc_auc_score, roc_curve, auc as sk_auc

# ── Paths ──────────────────────────────────────────────────────────
BASE_DIR   = '/content/mrnet_images/drive/MyDrive/MLII Final/Data/mrnet_images'
MASTER_CSV = '/content/drive/MyDrive/MLII Final/Data/unified_labels.csv'
SAVE_PATH  = '/content/drive/MyDrive/MLII Final/best_mri_transformer_v2.pth'

# ── Hyper-parameters ───────────────────────────────────────────────
BATCH_SIZE        = 2        # Physical batch (GPU memory constrained)
ACCUMULATION_STEPS = 8       # Effective batch = 16
EPOCHS            = 6       # More epochs — scheduler handles LR decay
MAX_SLICES        = 40       # Slices per view (centre-cropped / padded)
D_MODEL           = 512
N_HEAD            = 8
N_LAYERS          = 2
FOCAL_GAMMA       = 2.0      # Focal loss focus parameter
LR                = 3e-4     # Slightly higher than v1; cosine decay handles it
WEIGHT_DECAY      = 1e-2

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {DEVICE}")
```

```python
class MRNetDataset(Dataset):
    """
    Loads Axial / Coronal / Sagittal volumes and returns fixed-length
    tensors of shape (MAX_SLICES, 3, 224, 224) per view.
    Centre-crops or zero-pads the slice dimension so every sample has
    exactly MAX_SLICES slices.

    Path conventions (auto-detected):
      train : <base_dir>/axial/<pid>.npy
      valid : <base_dir>/valid_files/axial/<pid>.npy
    """
    # Subdirectory name used for each split inside BASE_DIR
    SPLIT_SUBDIR = {
        'train': '',             # files sit directly under base_dir/<view>/
        'valid': 'valid_files',  # files sit under base_dir/valid_files/<view>/
    }

    def __init__(self, base_dir, csv_path, split='train',
                 subset_ratio=1.0, max_slices=40, augment=False):
        self.base_dir   = base_dir
        self.split      = split
        self.max_slices = max_slices
        self.subdir     = self.SPLIT_SUBDIR[split]

        df = pd.read_csv(csv_path)
        self.labels = df[df['split'] == split].copy()
        self.labels['exam_id'] = self.labels['exam_id'].astype(str).str.zfill(4)

        if subset_ratio < 1.0 and split == 'train':
            self.labels = self.labels.sample(
                frac=subset_ratio, random_state=42).reset_index(drop=True)

        # Training augmentation (only applied when augment=True)
        aug_steps = [
            T.ToPILImage(),
            T.Resize((224, 224)),
            T.RandomHorizontalFlip(p=0.5),
            T.RandomRotation(degrees=10),
        ] if augment else [T.ToPILImage(), T.Resize((224, 224))]

        self.transform = T.Compose(aug_steps + [
            T.ToTensor(),
            T.Normalize(mean=[0.485, 0.456, 0.406],
                        std =[0.229, 0.224, 0.225])
        ])

    # ------------------------------------------------------------------
    def _build_path(self, view, pid):
        """
        Train: <base_dir>/<view>/<pid>.npy
        Valid: <base_dir>/valid_files/<view>/<pid>.npy
        """
        if self.subdir:
            return os.path.join(self.base_dir, self.subdir, view, f'{pid}.npy')
        return os.path.join(self.base_dir, view, f'{pid}.npy')

    # ------------------------------------------------------------------
    def _load_volume(self, path):
        vol = np.load(path).astype(np.float32)

        # Normalise to [0, 255] uint8 (required by ToPILImage)
        v_min, v_max = vol.min(), vol.max()
        if v_max - v_min > 1e-6:
            vol = (vol - v_min) / (v_max - v_min) * 255.0
        vol = vol.astype(np.uint8)

        s, h, w = vol.shape
        # Centre-crop or zero-pad to MAX_SLICES
        if s < self.max_slices:
            pad = np.zeros((self.max_slices - s, h, w), dtype=vol.dtype)
            vol = np.concatenate([vol, pad], axis=0)
        else:
            start = (s - self.max_slices) // 2
            vol   = vol[start : start + self.max_slices]

        slices = []
        for i in range(self.max_slices):
            rgb = np.stack([vol[i]] * 3, axis=-1)   # (H, W, 3)
            slices.append(self.transform(rgb))
        return torch.stack(slices)                   # (S, 3, 224, 224)

    # ------------------------------------------------------------------
    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        row = self.labels.iloc[idx]
        pid = row['exam_id']

        ax  = self._load_volume(self._build_path('axial',    pid))
        cor = self._load_volume(self._build_path('coronal',  pid))
        sag = self._load_volume(self._build_path('sagittal', pid))

        lbl = torch.tensor(
            [row['acl'], row['meniscus'], row['abnormal']],
            dtype=torch.float32)
        return ax, cor, sag, lbl
```

```python
# ── Build datasets & loaders ──────────────────────────────────────
train_dataset = MRNetDataset(BASE_DIR, MASTER_CSV, split='train',
                              max_slices=MAX_SLICES, augment=True)
valid_dataset = MRNetDataset(BASE_DIR, MASTER_CSV, split='valid',
                              max_slices=MAX_SLICES, augment=False)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE,
                          shuffle=True,  num_workers=2, pin_memory=True)
valid_loader = DataLoader(valid_dataset, batch_size=BATCH_SIZE,
                          shuffle=False, num_workers=2, pin_memory=True)

print(f"Train samples : {len(train_dataset)}")
print(f"Valid samples : {len(valid_dataset)}")

# Quick sanity check
ax, cor, sag, lbl = train_dataset[0]
print(f"Axial tensor  : {ax.shape}  dtype={ax.dtype}")
print(f"Labels        : {lbl}")
```

```python
# ── Path verification (run once to confirm paths are correct) ────
def _check_path(dataset, idx=0):
    row = dataset.labels.iloc[idx]
    pid = row['exam_id']
    for view in ['axial', 'coronal', 'sagittal']:
        p = dataset._build_path(view, pid)
        status = "✓ found" if os.path.exists(p) else "✗ MISSING"
        print(f"  {status}: {p}")

print("Train paths:")
_check_path(train_dataset)
print("Valid paths:")
_check_path(valid_dataset)
```

```python
class MultiViewTransformer(nn.Module):
    """
    Architecture:
      1. Shared ResNet-18 backbone (ImageNet init) extracts 512-d features
         per slice.
      2. Linear projection → d_model.
      3. Gaussian Positional Encoding: softly up-weights central slices.
      4. All-to-All Transformer Encoder over the concatenated sequence
         [axial | coronal | sagittal] (length = 3 * MAX_SLICES).
      5. Concatenated pooling (avg + max) → 2 * d_model feature vector.
      6. Three separate MLP heads (ACL / Meniscus / Abnormality).
    """

    def __init__(self, d_model=512, nhead=8, num_layers=2, dropout=0.1):
        super().__init__()

        # ── Backbone ──────────────────────────────────────────────
        resnet = resnet18(weights=ResNet18_Weights.IMAGENET1K_V1)
        self.backbone = nn.Sequential(*list(resnet.children())[:-1])
        # Output: (B*S, 512, 1, 1)  →  flatten to (B, S, 512)

        self.proj = nn.Linear(512, d_model)

        # ── Transformer ───────────────────────────────────────────
        enc_layer = nn.TransformerEncoderLayer(
            d_model=d_model, nhead=nhead,
            dim_feedforward=d_model * 4,
            dropout=dropout,
            batch_first=True,
            norm_first=True          # Pre-LN: more stable training
        )
        self.transformer = nn.TransformerEncoder(enc_layer, num_layers=num_layers)

        # ── Task heads ────────────────────────────────────────────
        # concat-pool produces 2*d_model; each head maps → 1 logit
        self.head_acl = self._make_head(d_model * 2)
        self.head_men = self._make_head(d_model * 2)
        self.head_abn = self._make_head(d_model * 2)

    # ------------------------------------------------------------------
    @staticmethod
    def _make_head(in_dim):
        return nn.Sequential(
            nn.Linear(in_dim, 256),
            nn.LayerNorm(256),
            nn.GELU(),
            nn.Dropout(0.4),
            nn.Linear(256, 64),
            nn.LayerNorm(64),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(64, 1)
        )

    # ------------------------------------------------------------------
    def freeze_bn(self):
        """Freeze all BatchNorm layers in the backbone."""
        for m in self.backbone.modules():
            if isinstance(m, nn.BatchNorm2d):
                m.eval()
                m.weight.requires_grad = False
                m.bias.requires_grad   = False

    # ------------------------------------------------------------------
    @staticmethod
    def _gaussian_prior(S: int, device):
        """Returns (1, S, 1) weight — peaks at the centre slice."""
        centre = S / 2.0
        x      = torch.arange(S, dtype=torch.float32, device=device)
        sigma  = S / 4.0
        prior  = torch.exp(-(x - centre) ** 2 / (2 * sigma ** 2))
        return prior.view(1, S, 1)

    # ------------------------------------------------------------------
    def _encode_view(self, x):
        """
        x : (B, S, 3, H, W)
        Returns (B, S, d_model) after backbone + projection + Gaussian prior.
        """
        B, S, C, H, W = x.shape
        flat = x.view(B * S, C, H, W)
        feat = self.backbone(flat).view(B, S, 512)        # (B, S, 512)
        feat = self.proj(feat)                             # (B, S, d_model)
        feat = feat * self._gaussian_prior(S, x.device)   # soft centre bias
        return feat

    # ------------------------------------------------------------------
    def forward(self, ax, cor, sag):
        f_ax  = self._encode_view(ax)    # (B, S, d)
        f_cor = self._encode_view(cor)
        f_sag = self._encode_view(sag)

        # All-to-all: single long sequence of length 3*S
        seq = torch.cat([f_ax, f_cor, f_sag], dim=1)   # (B, 3S, d)
        seq = self.transformer(seq)                      # (B, 3S, d)

        # Concatenated pooling
        f_avg  = seq.mean(dim=1)                         # (B, d)
        f_max  = seq.max(dim=1).values                   # (B, d)
        f_pool = torch.cat([f_avg, f_max], dim=1)        # (B, 2d)

        return self.head_acl(f_pool), self.head_men(f_pool), self.head_abn(f_pool)
```

```python
class FocalMultiTaskLoss(nn.Module):
    """
    Per-task Focal Loss with class-imbalance pos_weight.

    Focal Loss: FL(p_t) = -(1-p_t)^gamma * log(p_t)
      • gamma = 0  →  standard BCE
      • gamma = 2  →  hard-example mining (recommended)

    Task multipliers prioritise ACL (hardest / most clinically important).
    """
    TASK_MULTIPLIERS = [3.0, 2.0, 1.0]   # ACL, Meniscus, Abnormality

    def __init__(self, pos_weights: list, gamma: float = 2.0):
        super().__init__()
        self.gamma = gamma
        # Stored as buffer → moves to GPU with .to(device) automatically
        self.register_buffer(
            'pos_weights',
            torch.tensor(pos_weights, dtype=torch.float32))

    def forward(self, preds, targets):
        """
        preds   : tuple of 3 tensors, each (B, 1)
        targets : (B, 3)
        """
        total_loss = 0.0
        loss_parts = []

        for i in range(3):
            # FIX: use .squeeze(-1) instead of .squeeze() to be safe
            # when batch-size == 1 (bare .squeeze() would remove the batch dim too)
            logit  = preds[i].squeeze(-1)          # (B,)
            target = targets[:, i].float()          # (B,)

            # Weighted BCE (no reduction yet — we need per-sample values for focal)
            bce = nn.functional.binary_cross_entropy_with_logits(
                logit, target,
                pos_weight=self.pos_weights[i],
                reduction='none'
            )                                       # (B,)

            # Focal weight: (1 - p_t)^gamma
            p_t     = torch.exp(-bce)               # numerically stable
            f_loss  = ((1.0 - p_t) ** self.gamma * bce).mean()

            total_loss = total_loss + self.TASK_MULTIPLIERS[i] * f_loss
            loss_parts.append(f_loss.item())

        return total_loss, loss_parts
```

```python
# ── Class-imbalance weights (neg / pos ratio per task) ───────────
# Adjust these to match your actual dataset statistics.
# Formula: count_negative / count_positive for each label.
acl_pos_weight = 4.1   # ~4x more negatives than ACL positives
men_pos_weight = 1.8
abn_pos_weight = 0.5   # Abnormality is common, so weight is < 1

# ── Instantiate ───────────────────────────────────────────────────
model     = MultiViewTransformer(d_model=D_MODEL, nhead=N_HEAD,
                                  num_layers=N_LAYERS).to(DEVICE)
criterion = FocalMultiTaskLoss(
                pos_weights=[acl_pos_weight, men_pos_weight, abn_pos_weight],
                gamma=FOCAL_GAMMA
            ).to(DEVICE)   # .to(DEVICE) moves the registered buffer too

optimizer = optim.AdamW(model.parameters(), lr=LR,
                        weight_decay=WEIGHT_DECAY)

# ── Cosine LR Scheduler ───────────────────────────────────────────
# Decays LR smoothly from LR → ~0 over all epochs.
# T_max = effective number of gradient steps (approximation: use EPOCHS)
scheduler = optim.lr_scheduler.CosineAnnealingLR(
    optimizer, T_max=EPOCHS, eta_min=1e-6)

scaler = torch.amp.GradScaler('cuda')

print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
print(f"Optimizer LR    : {LR}")
print(f"Scheduler       : CosineAnnealingLR (T_max={EPOCHS}, eta_min=1e-6)")

# ── Optional: resume from checkpoint ─────────────────────────────
RESUME = False   # Set True to resume from SAVE_PATH
start_epoch   = 0
best_val_auc  = 0.0

if RESUME and os.path.exists(SAVE_PATH):
    ckpt = torch.load(SAVE_PATH, map_location=DEVICE)
    model.load_state_dict(ckpt['model_state_dict'])
    optimizer.load_state_dict(ckpt['optimizer_state_dict'])
    start_epoch  = ckpt['epoch'] + 1
    best_val_auc = ckpt['auc']
    print(f"Resumed from epoch {start_epoch} (best AUC={best_val_auc:.4f})")
```

```python
# =========================
# TRAIN FOR 3 EPOCHS ONLY
# =========================

EPOCHS = 3

# ── Metric helper ────────────────────────────────────────────────
def get_accuracy(logits, targets):
    """Binary accuracy from raw logits."""
    probs = torch.sigmoid(logits)
    preds = (probs > 0.5).float()
    return (preds == targets).float().mean().item()

# ── History ──────────────────────────────────────────────────────
history = {
    'train_loss': [], 'val_loss': [],
    'train_acl_acc': [], 'train_men_acc': [], 'train_abn_acc': [],
    'val_acl_acc':   [], 'val_men_acc':   [], 'val_abn_acc':   [],
    'val_acl_auc': [], 'val_men_auc': [], 'val_abn_auc': [],
    'lr': [],
}

task_keys = ['acl', 'men', 'abn']

# ── Main loop ────────────────────────────────────────────────────
for epoch in range(start_epoch, EPOCHS):

    # ── TRAIN ────────────────────────────────────────────────────
    model.train()
    model.freeze_bn()

    epoch_train_loss = 0.0
    t_accs = {j: [] for j in range(3)}
    optimizer.zero_grad()

    train_bar = tqdm(
        train_loader,
        desc=f"Ep {epoch+1}/{EPOCHS} [Train]",
        leave=False
    )

    for step, (ax, cor, sag, labels) in enumerate(train_bar):

        ax, cor, sag, labels = (
            t.to(DEVICE) for t in (ax, cor, sag, labels)
        )

        with torch.amp.autocast('cuda'):

            preds = model(ax, cor, sag)

            loss, loss_parts = criterion(preds, labels)

            scaled_loss = loss / ACCUMULATION_STEPS

        scaler.scale(scaled_loss).backward()

        if (step + 1) % ACCUMULATION_STEPS == 0:

            scaler.unscale_(optimizer)

            torch.nn.utils.clip_grad_norm_(
                model.parameters(),
                max_norm=1.0
            )

            scaler.step(optimizer)

            scaler.update()

            optimizer.zero_grad()

        epoch_train_loss += loss.item()

        for j in range(3):

            t_accs[j].append(
                get_accuracy(
                    preds[j].squeeze(-1),
                    labels[:, j]
                )
            )

        train_bar.set_postfix(
            loss=f"{loss.item():.3f}",
            acl=f"{loss_parts[0]:.3f}",
            men=f"{loss_parts[1]:.3f}",
            abn=f"{loss_parts[2]:.3f}"
        )

    # Flush remaining gradients
    if (len(train_loader)) % ACCUMULATION_STEPS != 0:

        scaler.unscale_(optimizer)

        torch.nn.utils.clip_grad_norm_(
            model.parameters(),
            max_norm=1.0
        )

        scaler.step(optimizer)

        scaler.update()

        optimizer.zero_grad()

    # Scheduler step
    scheduler.step()

    current_lr = scheduler.get_last_lr()[0]

    # ── VALIDATION ───────────────────────────────────────────────
    model.eval()

    epoch_val_loss = 0.0

    v_accs = {j: [] for j in range(3)}

    all_val_preds = [[] for _ in range(3)]

    all_val_true = [[] for _ in range(3)]

    with torch.no_grad():

        val_bar = tqdm(
            valid_loader,
            desc=f"Ep {epoch+1}/{EPOCHS} [Val]",
            leave=False
        )

        for ax, cor, sag, labels in val_bar:

            ax, cor, sag, labels = (
                t.to(DEVICE) for t in (ax, cor, sag, labels)
            )

            preds = model(ax, cor, sag)

            v_loss, _ = criterion(preds, labels)

            epoch_val_loss += v_loss.item()

            for j in range(3):

                logit = preds[j].squeeze(-1)

                v_accs[j].append(
                    get_accuracy(logit, labels[:, j])
                )

                probs = torch.sigmoid(logit)\
                    .cpu()\
                    .numpy()\
                    .flatten()\
                    .tolist()

                all_val_preds[j].extend(probs)

                all_val_true[j].extend(
                    labels[:, j]
                    .cpu()
                    .numpy()
                    .flatten()
                    .tolist()
                )

    # ── Metrics ──────────────────────────────────────────────────
    n_train = len(train_loader)

    n_val = len(valid_loader)

    history['train_loss'].append(
        epoch_train_loss / n_train
    )

    history['val_loss'].append(
        epoch_val_loss / n_val
    )

    history['lr'].append(current_lr)

    current_aucs = []

    for j, key in enumerate(task_keys):

        history[f'train_{key}_acc'].append(
            np.mean(t_accs[j])
        )

        history[f'val_{key}_acc'].append(
            np.mean(v_accs[j])
        )

        task_auc = roc_auc_score(
            all_val_true[j],
            all_val_preds[j]
        )

        history[f'val_{key}_auc'].append(task_auc)

        current_aucs.append(task_auc)

    avg_auc = float(np.mean(current_aucs))

    # ── Print Results ────────────────────────────────────────────
    print(f"\n── Epoch {epoch+1:02d}/{EPOCHS} ──  LR={current_lr:.2e}")

    print(
        f"  Loss  : train={history['train_loss'][-1]:.4f}  "
        f"val={history['val_loss'][-1]:.4f}"
    )

    for j, key in enumerate(task_keys):

        print(
            f"  {key.upper():3s}   : "
            f"acc={history[f'val_{key}_acc'][-1]:.4f}  "
            f"auc={current_aucs[j]:.4f}"
        )

    print(
        f"  Avg AUC: {avg_auc:.4f}  "
        f"(best={best_val_auc:.4f})"
    )

    # ── Save best model ──────────────────────────────────────────
    if avg_auc > best_val_auc:

        best_val_auc = avg_auc

        torch.save({

            'epoch': epoch,

            'model_state_dict': model.state_dict(),

            'optimizer_state_dict': optimizer.state_dict(),

            'scheduler_state_dict': scheduler.state_dict(),

            'auc': avg_auc,

            'history': history,

        }, SAVE_PATH)

        print(f"  ⭐ New best saved → {SAVE_PATH}")

print(f"\nTraining complete. Best Avg AUC = {best_val_auc:.4f}")
```

```python
def plot_results(history, all_val_true, all_val_preds):
    tasks = ['ACL', 'Meniscus', 'Abnormality']
    keys  = ['acl', 'men', 'abn']
    ep    = range(1, len(history['train_loss']) + 1)

    fig, axes = plt.subplots(4, 3, figsize=(18, 22))
    plt.subplots_adjust(hspace=0.38, wspace=0.3)

    for i, (task, key) in enumerate(zip(tasks, keys)):
        # Row 0 — Loss
        axes[0, i].plot(ep, history['train_loss'], 'b-o', ms=4, label='Train')
        axes[0, i].plot(ep, history['val_loss'],   'r-o', ms=4, label='Val')
        axes[0, i].set_title(f'Total Loss ({task} focus)')
        axes[0, i].set_xlabel('Epoch'); axes[0, i].set_ylabel('Loss')
        axes[0, i].legend(); axes[0, i].grid(True)

        # Row 1 — Accuracy
        axes[1, i].plot(ep, history[f'train_{key}_acc'], 'g-s', ms=4, label='Train')
        axes[1, i].plot(ep, history[f'val_{key}_acc'],   'm-s', ms=4, label='Val')
        axes[1, i].set_title(f'{task} Accuracy'); axes[1, i].set_xlabel('Epoch')
        axes[1, i].set_ylabel('Accuracy'); axes[1, i].set_ylim(0, 1)
        axes[1, i].legend(); axes[1, i].grid(True)

        # Row 2 — AUC over epochs
        axes[2, i].plot(ep, history[f'val_{key}_auc'], 'darkorange', marker='D',
                        ms=4, label='Val AUC')
        axes[2, i].axhline(max(history[f'val_{key}_auc']), color='grey',
                            linestyle='--', lw=1, label=f"Best={max(history[f'val_{key}_auc']):.3f}")
        axes[2, i].set_title(f'{task} AUC per Epoch'); axes[2, i].set_xlabel('Epoch')
        axes[2, i].set_ylabel('AUC'); axes[2, i].set_ylim(0.4, 1.0)
        axes[2, i].legend(); axes[2, i].grid(True)

        # Row 3 — ROC Curve (final epoch)
        fpr, tpr, _ = roc_curve(all_val_true[i], all_val_preds[i])
        area = sk_auc(fpr, tpr)
        axes[3, i].plot(fpr, tpr, 'darkorange', lw=2, label=f'AUC={area:.3f}')
        axes[3, i].plot([0,1],[0,1], 'navy', lw=1.5, linestyle='--')
        axes[3, i].set_xlim([0,1]); axes[3, i].set_ylim([0,1.02])
        axes[3, i].set_title(f'{task} ROC Curve')
        axes[3, i].set_xlabel('FPR'); axes[3, i].set_ylabel('TPR')
        axes[3, i].legend(loc='lower right'); axes[3, i].grid(True)

    plt.suptitle('MRNet Multi-View Transformer v2 — Training Results',
                 fontsize=18, y=0.995, fontweight='bold')
    plt.savefig('/content/drive/MyDrive/MLII Final/training_results_v2.png',
                dpi=120, bbox_inches='tight')
    plt.show()
    print("Figure saved to Drive.")

plot_results(history, all_val_true, all_val_preds)
```
